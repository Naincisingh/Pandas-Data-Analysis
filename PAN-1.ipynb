```python
"""QTS.5"""
##1.
"""DataFrame.size:
The DataFrame.size attribute returns the total number of elements in the DataFrame, which is equal to the product of the number of rows and the number of columns.
It gives you the total count of values in the DataFrame, regardless of their content.
The value returned by DataFrame.size is an integer.
Example:"""
import pandas as pd

data = {"A": [1, 2, 3],
        "B": [4, 5, 6]}
df = pd.DataFrame(data)

print(df.size)  # Output: 6

##2.
"""DataFrame.shape:
The DataFrame.shape attribute returns a tuple representing the dimensions of the DataFrame, where the first element is the number of rows and the second element is the number of columns.
It gives you the structure of the DataFrame in terms of rows and columns.
The values returned by DataFrame.shape are integers.
Example:"""
import pandas as pd

data = {"A": [1, 2, 3],
        "B": [4, 5, 6]}
df = pd.DataFrame(data)

print(df.shape)  # Output: (3, 2)
```

```python
""""QTS.7"""
import pandas as pd
data={"email":['sudh.kumar@gmail.com','nainci@gmail.com','john.doe@gmail.com']}
df=pd.DataFrame(data)
df

def extract(x):   
        if x.find('@')!=False:
            return x[:x.find('@')]
        else:
            return "not present"

        
df['username']=df['email'].apply(extract)     
df
```

```python
"""QTS.8"""
import pandas as pd

def select_rows(df):
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]
    return selected_rows

# Example data
data = {"A": [3, 8, 6, 2, 9],
        "B": [5, 2, 9, 3, 1],
        "C": [1, 7, 4, 5, 2]}

df = pd.DataFrame(data)

# Call the function to select rows
selected_df = select_rows(df)

print(selected_df)
```
